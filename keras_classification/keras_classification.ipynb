Model Preparation

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

from tensorflow.keras.models import  Sequential
from tensorflow.keras.layers import  Dense, Dropout
```

```python
X = data.drop('benign_0__mal_1', axis=1).values
y = data['benign_0__mal_1'].values
```

```python
X, X.shape
```

```python
y, y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=101)
```

```python
scalar = MinMaxScaler()
```

```python
X_train = scalar.fit_transform(X_train)
X_test = scalar.transform(X_test)
```

Model 1

```python
model = Sequential()

model.add(Dense(30, activation='relu'))
model.add(Dense(15, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.compile(optimizer='adam', loss='binary_crossentropy')
```

```python
model.fit(x=X_train, y=y_train, epochs=500, validation_data=(X_test, y_test))
```

```python
losses = pd.DataFrame(model.history.history)
```

```python
losses
```

```python
losses.plot()
```

Model 2

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
model = Sequential()

model.add(Dense(30, activation='relu'))
model.add(Dense(15, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.compile(optimizer='adam', loss='binary_crossentropy')
```

```python
early_stop = EarlyStopping(monitor='val_loss', mode='min', patience=20, verbose=1)
```

```python
model.fit(x=X_train, y=y_train, epochs=500, validation_data=(X_test, y_test), callbacks=early_stop)
```

```python
losses = pd.DataFrame(model.history.history)
```

```python
losses.plot()
```

Model 3

```python
model = Sequential()

model.add(Dense(30, activation='relu'))
model.add(Dropout(0.5))

model.add(Dense(15, activation='relu'))
model.add(Dropout(0.5))

model.add(Dense(1, activation='sigmoid'))

model.compile(optimizer='adam', loss='binary_crossentropy')
```

```python
model.fit(x=X_train, y=y_train, epochs=600, validation_data=(X_test, y_test), callbacks=early_stop)
```

```python
losses = pd.DataFrame(model.history.history)
```

```python
losses.plot()
```

Model Analysis

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
predictions = model.predict(X_test)
```

```python
predictions = (predictions>0.5).astype('int32')
```

```python
print(classification_report(y_test, predictions))
```

```python
confusion_matrix(y_test, predictions)
```
